```python
# using recursive find the sum of this series 1 + 2 + 3 + 4 + 5

def total(start,end):
    if start>end:
        return 0
    else:
        return start + total(start+1,end)

total(1,5)
```

```python
#  1^2 + 2^2 + 3^2 + 4^2 + 5^2


def sq(num):
    return num**2


def total(start,end):
    if start>end:
        return 0
    else:
        return sq(start) + total(start+1,end)

total(1,10)
```

```python
class Employee:
    def __init__(self,empid,name,job,sal):
        self.empid = empid
        self.name = name             # constructor: special method that generally used to initialize instance variable
        self.job = job
        self.sal = sal

    def showDetails(self):
        print(self.empid,self.name,self.job,self.sal)

emplist = []

while True:
    print("\n1 Add Record \n2 Delete Record \n3 Display Records \n4 Exit:")
    option = int(input("Enter your option:"))
    if option==1:
        eid = int(input("Enter empid:"))
        name = input("Enter name:")
        job = input("Enter job:")
        sal = float(input("Enter salary"))
        emplist.append(Employee(eid,name,job,sal))
        print("Record added successfully..")

    if option==2:
        index = int(input("which record you want to delete:"))
        if len(emplist)<index:
            print(f'sorry.. {len(emplist)} records are present')
        else:
            emplist.pop(index-1)
            print("Record deleted successfully")

    if option==3:
        for item in emplist:
            item.showDetails()

    if option==4:
        print("Thank you")
        break
```

```python
class Account:
    
    def __init__(self,amt):         # self : current object
        self.__mybal = amt

    
    def getBalance(self):          # Getter function 
        return self.__mybal


    def setIntrest(self):         # setter function
        self.__mybal = self.__mybal + self.__mybal * .05


obj = Account(8000)

ac2 = Account(6000)

print(ac2.getBalance())

print(obj.getBalance())

obj.setIntrest()
ac2.setIntrest()

print(ac2.getBalance())
print(obj.getBalance())
```

```python
class Employee:
    def __init__(self,eid,name,job,sal,loc):
        self.eid = eid
        self.name = name
        self.job = job
        self.sal = sal
        self.loc = loc

    def showDetails(self):
        print(self.eid,self.name,self.job,self.sal,self.loc)

    def getName(self):
        return self.name

    def setSalary(self,amt):
        self.sal = amt
    


emplist = []

emplist.append(Employee(555,"Shekar","programmer",65321.85,"gachibowli"))
emplist.append(Employee(999,"Rohit","manager",85324.85,"hitech_city"))
emplist.append(Employee(111,"Amith","analyst",71235.85,"gachibowli"))
emplist.append(Employee(333,"Raju","teamlead",69251.85,"gachibowli"))
emplist.append(Employee(777,"Lokesh","programmer",63251.85,"hitech_city"))


name = input("Enter name:")

for emp in emplist:
    if emp.getName()==name:
        print("found")
        print(emp.name,emp.eid,emp.job,emp.sal)
        break
else:
    print("Not found")


name = input("Enter name:")

for emp in emplist:
    if emp.getName()==name:
        salary = int(input("Enter salary:"))
        emp.setSalary(salary)
        print("updated successfully")
        break
else:
    print("Not found")


name = input("Enter name:")

for emp in emplist:
    if emp.getName()==name:
        print("found")
        print(emp.name,emp.eid,emp.job,emp.sal)
        break
else:
    print("Not found")

```

```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

class Employee(Person):
    def __init__(self,name,age,eid,job,sal):
        super().__init__(name,age)
        self.eid = eid
        self.job = job
        self.sal = sal

    def showDetails(self):
        print(self.name,self.age,self.eid,self.job,self.sal)
        
e = Employee("shekar",25,555,"analyst",65324.23)
e.showDetails()
```

```python
class GParent:
    def GTalent(self):
        print("he is a writer")

class Parent(GParent):
    def PTalent(self):
        print("he is a musician")

class Child(Parent):
    def CTalent(self):
        print("he is an actor")

c  = Child()

c.CTalent()
c.PTalent()
c.GTalent()

```

```python
class Complex:                       #  5 + 3j
    def __init__(self,real,img):
        self.real = real
        self.img = img

    def display(self):
        print(self.real,"+",self.img,"j")

    def __add__(self,c):
        self.real = self.real + c.real
        self.img = self.img + c.img
        
c1 = Complex(5,3)
c2 = Complex(7,5)

c1.display()
c2.display()
c1 + c2            # c1.add(c2)

c1.display()
```

```python
# overriding:

class Parent:
    def abc(self):
        print("abc of parent")

class Child(Parent):
    def abc(self):
        print("abc of child")

c = Child()
c.abc()
```
